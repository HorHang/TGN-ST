# Plot of Activation Function and Its Gradient

```python
from numpy import *
import math
import matplotlib.pyplot as plt
import torch

t = linspace(-5, 5, 10000)
sigmoid = torch.sigmoid(torch.tensor(t)).numpy()
tanh = torch.tanh(torch.tensor(t)).numpy()
relu = torch.relu(torch.tensor(t)).numpy()
gelu = torch.nn.functional.gelu(torch.tensor(t)).numpy()
softplus = torch.nn.functional.softplus(torch.tensor(t)).numpy()
leaky_relu = torch.nn.functional.leaky_relu(torch.tensor(t), 0.1).numpy()

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(t, sigmoid, 'r', label='Sigmoid')
ax.plot(t, tanh, 'b', label='Tanh')
ax.plot(t, relu, 'g', label='ReLU')
ax.plot(t, gelu, 'y', label='GELU')
ax.plot(t, leaky_relu, 'c', label='Leaky ReLU')
ax.plot(t, softplus, 'm', label='Softplus')

plt.legend()
plt.ylim(ymin=-1.2, ymax=3)
plt.xlabel('x')
plt.ylabel('$\sigma$(x)')
plt.title('Activation Functions')
plt.grid(True)
plt.savefig("activation_functions.png", dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

def relu(x):
  return np.maximum(0, x)

def relu_grad(x):
  return np.where(x > 0, 1, 0)  # 1 for positive values, 0 for negative

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def sigmoid_grad(x):
  return sigmoid(x) * (1 - sigmoid(x))

def tanh(x):
  return np.tanh(x)

def tanh_grad(x):
  return 1 - tanh(x) ** 2

def leaky_relu(x, alpha=0.01):
  return np.maximum(alpha * x, x)

def leaky_relu_grad(x, alpha=0.01):
  return np.where(x > 0, 1, alpha)


def softplus(x):
  return np.log(1 + np.exp(x))

def softplus_grad(x):
  return 1 / (1 + np.exp(-x))

def gelu(x):
  return x * erf(x / np.sqrt(2)) + 0.5 * x * np.abs(x) * (1 - erf(x / np.sqrt(2)))

def gelu_grad(x):
  cdf = erf(x / np.sqrt(2))
  pdf = np.exp(-x**2 / 2) / (np.sqrt(2 * np.pi))
  return cdf + x * pdf + 0.5 * (1 - x * np.abs(x)) * pdf


# Define the x-axis values
x = np.linspace(-5, 5, 100)

# Plot the activation functions and their gradients

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, sigmoid_grad(x), 'r', label='Sigmoid Gradient')
ax.plot(x, tanh_grad(x), 'b', label='Tanh Gradient')
ax.plot(x, relu_grad(x), 'g', label='ReLU Gradient')
ax.plot(x, gelu_grad(x), 'y', label='ReLU')
ax.plot(x, leaky_relu_grad(x), 'c', label='Leaky ReLU Gradient')
ax.plot(x, softplus_grad(x), 'm', label='Softplus Gradient')

# Customize the plot
plt.xlabel('x')
plt.ylabel('Gradient of $\sigma$(x)')
plt.title('Gradients of Activation Functions')
plt.legend()
plt.grid(True)
plt.savefig("activation_functions_gradients.png", dpi=300)
plt.show()
```

# Example of Early Stopping

```python
import matplotlib.pyplot as plt

# Sample training and validation loss data (replace with actual data from your training)
train_loss = [2.5, 2.2, 2.0, 1.8, 1.7, 1.65, 1.6, 1.58, 1.57, 1.56, 1.55, 1.50]
val_loss = [2.7, 2.4, 2.1, 1.9, 1.85, 1.8, 1.78, 1.75, 1.73, 1.8, 1.85, 2.00]  # Simulates validation loss increase

# Set patience (number of epochs with no improvement) for early stopping
patience = 3

# Identify the epoch where early stopping occurred (assuming it did)
min_val_loss = min(val_loss)
early_stopping_epoch = val_loss.index(min_val_loss)  # Find index of minimum value
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the training and validation loss
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')

# Highlight the epoch of early stopping (if applicable)
if early_stopping_epoch < len(val_loss) - patience:  # Check if stopping occurred before patience limit
  plt.axvline(x=early_stopping_epoch, color='red', linestyle='--', label='Early Stopping')

# Customize the plot
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss with Early Stopping')
plt.legend()
plt.grid(True)
plt.savefig("early_stopping.png", dpi=300)
plt.show()
```

# Example of Networkx Graph Creation

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add two nodes
G.add_edge(1, 2)  # Add a directed edge from A to B
G.add_edge(2, 3)  # Add a directed edge from B to C
G.add_edge(3, 1)
G.add_edge(1, 3)
G.add_node(4)

# Set positions for the nodes (optional, but useful for visualization)
pos = {1: (0.5, 1), 2: (1, 2), 3: (1.5, 0), 4: (2, 1)}

fig, ax = plt.subplots(figsize=(10, 5))
# Draw the graph
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_edges(G, pos, arrows=True)  # Add arrows for directed edges

# Display labels for nodes
nx.draw_networkx_labels(G, pos, font_size=12)

# Show the plot
plt.title('Sample Directed Graph')
plt.axis('off')  # Hide unnecessary axes
plt.savefig("sample_node.png", dpi=300)
plt.show()
```
